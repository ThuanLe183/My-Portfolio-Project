## Data cleaning in Pandas

```python
import pandas as pd
```

```python
df4 = pd.read_excel(r"C:\Users\admin\OneDrive\Desktop\My Portfolio Project\Customer Call List (1).xlsx")
```

```python
df4
```

```python
df4 = df4.drop_duplicates()
```

```python
df4
```

```python
df4 = df4.drop (columns = 'Not_Useful_Column')
```

```python
df4
```

```python
df4['Last_Name'] = df4['Last_Name'].str.strip('123._/')
```

```python
df4
```

```python
df4['Phone_Number'] = df4['Phone_Number'].str.replace('[^a-zA-Z0-9]', '', regex=True)
```

```python
df4
```

```python
df4["Phone_Number"] = df4["Phone_Number"].apply(lambda x: str(x))
```

```python
df4["Phone_Number"] = df4["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
```

```python
df4
```

```python
df4["Phone_Number"] = df4["Phone_Number"].str.replace('nan--','')
df4["Phone_Number"] = df4["Phone_Number"].str.replace('Na--','')
df4
```

```python
df4[["Street_Address", "State", "Zip_Code"]] = df4["Address"].str.split(',', n=2,  expand=True)
```

```python
df4
```

```python
df4["Do_Not_Contact"] = df4["Do_Not_Contact"].str.replace('Yes','Y')

df4["Do_Not_Contact"] = df4["Do_Not_Contact"].str.replace('No','N')
df4
```

```python
df4["Paying Customer"] = df4["Paying Customer"].str.replace('Yes','Y')

df4["Paying Customer"] = df4["Paying Customer"].str.replace('No','N')
df4
```

```python
for x in df4.index:
    if df4.loc[x, "Phone_Number"] == '':
        df4.drop(x, inplace=True)

df4
```

```python
df4 = df4.reset_index(drop=True)
df4
```
